**Keras**: Keras is a high-level deep learning library that runs on top of TensorFlow, allowing users to build and train neural networks with ease. It provides a user-friendly API and supports both convolutional and recurrent networks.

Keras is a high-level deep learning library that runs on top of TensorFlow (or other backend engines). Here are some basic Keras code examples:

Sequential Model for Binary Classification:

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# Generate some toy data for binary classification
np.random.seed(42)
X = np.random.rand(100, 2)  # 100 samples with 2 features each
y = (X[:, 0] + X[:, 1] > 1).astype(int)  # Binary labels based on a simple condition

# Create the neural network model
model = Sequential()
model.add(Dense(64, input_dim=2, activation='relu'))  # 64 neurons in the first layer, input_dim=2 for 2 features
model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron and sigmoid activation for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=50, batch_size=16, verbose=1)

# Make predictions on new data
new_data = np.array([[0.8, 0.3], [0.2, 0.9]])
predictions = model.predict(new_data)
print(predictions)
```

Sequential Model for Multiclass Classification:

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical

# Generate some toy data for multiclass classification
np.random.seed(42)
X = np.random.rand(100, 2)  # 100 samples with 2 features each
y = np.random.randint(0, 3, size=100)  # Three classes (0, 1, 2)

# Convert labels to one-hot encoded format
y_one_hot = to_categorical(y, num_classes=3)

# Create the neural network model
model = Sequential()
model.add(Dense(64, input_dim=2, activation='relu'))  # 64 neurons in the first layer, input_dim=2 for 2 features
model.add(Dense(3, activation='softmax'))  # Output layer with 3 neurons and softmax activation for multiclass classification

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y_one_hot, epochs=50, batch_size=16, verbose=1)

# Make predictions on new data
new_data = np.array([[0.8, 0.3], [0.2, 0.9]])
predictions = model.predict(new_data)
print(predictions)
```

Sequential Model for Regression:

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense

# Generate some toy data for regression
np.random.seed(42)
X = np.random.rand(100, 1)
y = 3 * X + 2 + 0.1 * np.random.randn(100, 1)

# Create the neural network model
model = Sequential()
model.add(Dense(64, input_dim=1, activation='relu'))  # 64 neurons in the first layer, input_dim=1 for 1 feature
model.add(Dense(1))  # Output layer with 1 neuron for regression

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, y, epochs=50, batch_size=16, verbose=1)

# Make predictions on new data
new_data = np.array([[0.8], [0.2]])
predictions = model.predict(new_data)
print(predictions)
```
